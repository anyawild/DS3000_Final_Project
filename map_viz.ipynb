```python
import pandas as pd
import plotly.express as px
```

```python
euro_c_df_clean = pd.read_csv('euro_c_df_clean.csv')
euro_c_df_clean.head()
```

```python
# group by lat and long
df_grouped = euro_c_df_clean[['latitude', 'longitude','bright_ti5']].groupby(['latitude', 'longitude']).sum('bright_ti5')
df_grouped = df_grouped.reset_index()
fig = px.scatter_geo(
    df_grouped,
    lat='latitude',
    lon='longitude',
    color='bright_ti5',
    hover_data = ['bright_ti5'],
)
# start looking at europe
fig.update_layout(
    geo=dict(
        center={"lat": 54, "lon": 13}, 
        projection_scale=4
    )
)
```
